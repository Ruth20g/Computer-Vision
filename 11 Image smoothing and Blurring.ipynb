### Averaging Kernel

```python
import cv2
import numpy as np
```

```python
# Load an image
image = cv2.imread("images/lenna.png")

# Define a 3x3 averaging kernel
kernel = np.ones((5, 5), np.float32) / 25

# Apply the kernel to smooth the image
smoothed_image = cv2.filter2D(image, -1, kernel)

# Show the original and smoothed images
cv2.imshow("Original",image)
cv2.imshow("Smoothed",smoothed_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# to display both images together
new_image = np.hstack([image, smoothed_image])
```

```python
cv2.imshow("Both Images",new_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Gaussian Kernel

```python
kernel_size = 7   # adjust as needed
sigma = 0.35 * ((kernel_size - 1) * 0.5 - 1) + 0.9     # Gaussian Standard Deviation
kernel = cv2.getGaussianKernel(kernel_size, sigma)
```

```python
sigma
```

```python
kernel
```

```python
# apply the kernel to smooth the image
smoothed_image = cv2.filter2D(image, -1, kernel)
# Show the original and smoothed images
cv2.imshow("Original",image)
cv2.imshow("Smoothed",smoothed_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# to display both images together
new_image = np.hstack([image, smoothed_image])
cv2.imshow("Both Images",new_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Median Filter

```python
gray = cv2.imread('images/lenna.png', 0)
```

```python
gray.shape
```

```python
import matplotlib.pyplot as plt
```

```python
kernel_size = 15
median = cv2.medianBlur(gray, kernel_size)
plt.figure(figsize=(10,6))
plt.subplot(121), plt.imshow(gray, cmap='Blues'), plt.title('Original Image')
plt.subplot(122), plt.imshow(median, cmap='gray'), plt.title('Median filtered Image')
```

```python
image = cv2.imread("images/lenna.png")

# Define the coordinates of the patch
x1, y1 = 100, 100
x2, y2 = 300, 300

# Extract the patch from original image
patch = image[x1:y1,x2:y2]
blurred_image = cv2.medianBlur(patch,kernel_size=20)
```
